Suppose you have a function 
$$
f(x) = 2x
$$
and you're looking to plot this function over the interval $[0,10]$. How would you go about doing this?

First we define the function using Python's function definition:
```Python
def f(x):
    return 2*x
```
Now we can call our function `f(x)` by inputting any value of `x` we want.


````python
def f(x):
    return 2*x

f(5)
````

Next to plot the function, we take advantage of a Python package called `matplotlib`.
Firstly we make a domain for `x` over the interval $[0,10]$. Since code is discrete and not continious, the interval we make will be discrete, but by making the spacing between values smaller and smaller, we can approximate how the continious plot would look like.

We use the `numpy` package to make a linear interval:
-  used to create an array of evenly spaced values over a specified range.
```python
import numpy as np
numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)
```
-`start`: The starting value of the sequence.

-`stop`: The end value of the sequence.

-`num` (optional): The number of evenly spaced samples to generate. The default value is 50.

-`endpoint` (optional): If True (default), the stop value is included in the sequence. If False, the stop value is not included.

-`retstep` (optional): If True, it returns a tuple with the array of values and the spacing between them. The default is False.

-`dtype` (optional): The data type of the output array. If not specified, it is inferred from the input values

This code will make an array `x`, which has 11 evenly spaced values between $0$ and $10$. 

````python
x=np.linspace(0,10,11)
````

Now, we can make a table for $f(x)$:

````python
x=np.linspace(0,10,11)
y=f(x)

def graph(x,y):

    plt.plot(x,y)
    # Add horizontal and vertical lines at the zero points
    plt.ylabel('f(x)')
    plt.xlabel('x')
    plt.axis('tight')
    plt.grid(True, linestyle="--", alpha=0.6)

    plt.axhline(0, color='black', linestyle='-', linewidth=1)
    plt.axvline(0, color='black', linestyle='-', linewidth=1)

    plt.show()
   
graph(x,y)
````
