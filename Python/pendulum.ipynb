
## Simulating a Simple Pendulum

A simple pendulum consists of a mass (bob) attached to the end of a string or rod of fixed length. It can oscillate back and forth under the influence of gravity. In this notebook, we will simulate the motion of a simple pendulum and visualize its behavior.

### Importing Libraries

Let's start by importing the necessary libraries for our simulation:

Pendulum Parameters
We need to define some parameters for our pendulum:

Length of the pendulum arm (in meters)
Initial angle (in radians)
Initial angular velocity (in radians per second)
Time span for simulation (in seconds)

Length of the pendulum arm (in meters)

Initial angular velocity (in radians per second)

```python
# Pendulum parameters
length = 1.0  # Length of the pendulum arm (meters)
initial_angle = np.pi / 4  # Initial angle (45 degrees in radians)
initial_angular_velocity = 0.0  # Initial angular velocity (rad/s)

# Time span for simulation
t_start = 0.0
t_end = 10.0
num_points = 1000
t = np.linspace(t_start, t_end, num_points)
```

Differential Equation for Pendulum Motion
The motion of a simple pendulum can be described by a second-order ordinary differential equation (ODE). We will define the ODE and create a function to represent it:

```python
def pendulum_ode(y, t):
    """
    Differential equation for a simple pendulum.
    y[0] is the angular displacement.
    y[1] is the angular velocity.
    """
    g = 9.81  # Acceleration due to gravity (m/s^2)
    theta, omega = y
    dydt = [omega, -g / length * np.sin(theta)]
    return dydt
```

Solving the Differential Equation
We will use the odeint function from SciPy to solve the differential equation and simulate the pendulum's motion:

```python
# Initial conditions
initial_conditions = [initial_angle, initial_angular_velocity]

# Solve the ODE
pendulum_solution = odeint(pendulum_ode, initial_conditions, t)

# Extract angular displacement and angular velocity
angular_displacement = pendulum_solution[:, 0]
angular_velocity = pendulum_solution[:, 1]
```

Visualizing the Pendulum Motion
Let's plot the angular displacement of the pendulum as a function of time to visualize its motion:

```python
# Plot the pendulum motion
plt.figure(figsize=(10, 4))
plt.plot(t, angular_displacement, label="Angular Displacement")
plt.xlabel("Time (s)")
plt.ylabel("Angular Displacement (rad)")
plt.title("Simple Pendulum Motion")
plt.grid(True)
plt.legend()
plt.show()
```
