```python
import numpy as np
import matplotlib.pyplot as plt
import mercury as mr
```

```python
a = mr.Slider(label="a", min=-5, max=5, value=1)
b = mr.Slider(label="b", min=-5, max=5, value=1)
c = mr.Slider(label="c", min=-5, max=5, value=1)
```

We want to turn 
$$ AX+BY=C$$ into $$y=mx+b$$
to do this:
We rearrange:
$$
BY=C-AX \\\
Y=(C-AX)/B  \\\
Y = \frac{-A}{B}x + \frac{C}{B} \\\
$$
therefore,
$m=-\frac{A}{B}$ and $b=\frac{C}{B}$

```python
def vertex(a, b, c):
    h=-b/(2*a)
    k = c-(b**2)/(4*a)
    return h,k

def f(x,a,b,c):
    h,k = vertex(a,b,c)
    return (x-h)**2 + k 

def quadratic(a,b,c):
    h,k = vertex(a,b,c)
    x=np.arange(-5,5,0.1)
    y=f(x,a,b,c)
    plt.plot(x,y)
    plt.plot(h,k, 'r+')
    plt.ylabel('some numbers')
    plt.xlabel('guess what this does')
    plt.axis('tight')
    plt.grid(True)
    plt.show()

quadratic(a,b,c)
```
